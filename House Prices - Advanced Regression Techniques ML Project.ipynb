```python
# detecting outliers in df_num
def outliers_iden(df1):
    df = df1.copy()
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3-q1
    
    lower_bound = q1 - (1.5*iqr)
    upper_bound = q3 + (1.5*iqr)
    outliers_list = []
    for n,c in df.items():
        df[n] = c.mask(c < lower_bound[n], lower_bound[n])
        df[n] = c.mask(c > upper_bound[n], upper_bound[n])
        if c.any() > upper_bound[n] or c.any() < lower_bound[n]:
            outliers_list.append(n)
            
    return (outliers_list, df)

(outliers_list, df_num1) = outliers_iden(df_num)
        
        
```
